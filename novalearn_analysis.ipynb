```python
import pandas as pd
import matplotlib.pyplot as plt


# dataset with products in English
data = {
    'Product': ['Online Course', 'Software Subscription', 'E-book', 'Mentorship', 'Paid Webinar'],
    'Sales': [5200, 4600, 1800, 3100, 2400]
}

df = pd.DataFrame(data)
df

# Pie chart showing sales distribution
plt.figure(figsize=(6,6))  # Square shape

plt.pie(
    df['Sales'],
    labels=df['Product'],
    autopct='%1.1f%%',           # Show percentages (1 decimal place)
    colors=plt.cm.Paired.colors  # Predefined color palette
)

plt.title('Sales Distribution by Product')
plt.tight_layout()
plt.show()

```
